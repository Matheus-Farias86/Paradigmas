**Exercício de vetores 2**
1. Faça um programa em Python que receba do usuario dois vetores, A e B, com 10 números inteiros cada. Crie um novo vetor denominado C calculando C = A - B. Mostre na tela os dados do vetor C.

```python
def main():
  # Declarando vetores
  vetor_a = []
  vetor_b = []
  vetor_c = []
  # Recebendo os dados dos vetores
  for i in range(10):
    vetor_a.append(int(input("Digite os elementos %d do vetor A: " % i)))
    vetor_b.append(int(input("Digite os elementos %d do vetor B: " % i)))

  for i in range(10):
    vetor_c.append(vetor_a[i] - vetor_b[i])

  print("Vetor C: ")
  for i in range(10):
    print("Elemento %d do vetor C: %d" % (i, vetor_c[i]))
main()
```
