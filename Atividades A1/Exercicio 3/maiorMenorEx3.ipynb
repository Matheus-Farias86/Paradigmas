```python
def main():
  # Cria um vetor de 10 elementos inteiros
  vetor = [5, 12, 7, 20, 15, 8, 3, 11, 6, 9]

  # Calcula a soma, o maior valor e o menor valor
  soma = 0
  maior_valor = vetor[0]
  menor_valor = vetor[0]

  for valor in vetor:
    soma += valor
    if valor > maior_valor:
      maior_valor = valor
    if valor < menor_valor:
      menor_valor = valor

  media = soma / len(vetor)

  print("Média:", media)
  print("Maior valor:", maior_valor)
  print("Menor valor:", menor_valor)

main()
```
