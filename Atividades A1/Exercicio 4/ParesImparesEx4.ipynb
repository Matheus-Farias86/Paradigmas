```python
def main():
    # Inicializando variáveis
    numbers = []
    sumEven = 0
    countOdd = 0

    # Receber os números do usuário
    print("Digite 6 números inteiros: ")
    for i in range(6):
        number = int(input())
        numbers.append(number)

    # Mostrar números pares e calcular a soma
    print("Números pares digitados: ")
    for num in numbers:
        if num % 2 == 0:
            print(num)
            sumEven += num

    print("Soma dos números pares:", sumEven)

    # Mostrar números ímpares e contar a quantidade
    print("Números ímpares digitados: ")
    for num in numbers:
        if num % 2 != 0:
            print(num)
            countOdd += 1

    print("Quantidade de números ímpares:", countOdd)
main()
```
